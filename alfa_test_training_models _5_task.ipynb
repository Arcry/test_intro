# Моделирование 

## Построение моделей 

Воспользуемся алгоритмами  RandomForestClassifier, SVC, GradientBoostingClassifier и LogisticRegression и CatBoost.

```python
# параметры алгоритмов
gbc_params = {'learning_rate': np.arange(0.05, 0.15, 0.01),
             'n_estimators': np.arange(50, 200, 50)} # GradientBoostingClassifier

rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}

svc_params = {'kernel': ['linear', 'rbf'], # SVC
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.1, 1, 0.1)}

cat_params = {'iterations': [100, 250], 
          'learning_rate': np.arange(0.05, 0.1, 0.01), 
          'depth': [2, 3]
         }

skf = StratifiedKFold(n_splits=5, random_state=17, shuffle=True)
gbc = GradientBoostingClassifier(random_state=17)
rfc = RandomForestClassifier(random_state=17)
svc = SVC(random_state=17)
lr = LogisticRegression(random_state=17)
cat = CatBoostClassifier(random_state=17,
                         loss_function='Logloss', 
                         )

models = [gbc, rfc, svc, lr, cat]
parametrs = [gbc_params, rfc_params, svc_params, lr_params, cat_params]
```

Обучим все модели

```python
models_CV = []
for model, param in zip(models, parametrs):
    models_CV.append(GridSearchCV(model, 
                         param, 
                         cv=skf).fit(X_train, y_train))
```

 Выведем лучшие параметры обученных моделей

```python
for model in models_CV:
    print(f'Best params of {type(model.estimator).__name__} are {model.best_params_}')
```

Так классы несбалансированы в качестве метрик будут использоваться ROC_AUC, Recall, Precision, F1 score и Confusion matrix

```python
for model in models_CV:
    print(f'ROC_AUC of {type(model.estimator).__name__} is {roc_auc_score(y_val, model.predict(X_val)):.3f}')
    print(f'Recall of {type(model.estimator).__name__} is {recall_score(y_val, model.predict(X_val)):.3f}')
    print(f'Precision of {type(model.estimator).__name__} is {precision_score(y_val, model.predict(X_val)):.3f}')
    print(f'F1 score of {type(model.estimator).__name__} is {f1_score(y_val, model.predict(X_val)):.3f}')
    print('Confusion matrix:')
    print(confusion_matrix(y_val, model.predict(X_val)))
    print()
```
